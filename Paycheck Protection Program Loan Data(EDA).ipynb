# Conclusion:

1) LoanRange: There seems to be a clear distribution of loans based on their amounts, with most loans falling in the range of  150,000 𝑡𝑜 350,000.

2) BusinessName: It seems there are some generic names like "NEW APPLICATION" that might need further investigation or cleaning.

3) Address: Some addresses appear to be repetitive or not well-formed, such as "PO BOX" and "10 MIDDLE ST 17th Floor". There might be inconsistencies or discrepancies in address formatting.

4) City: There are different city names, but some are in uppercase and others in title case. Standardizing the case might be necessary.

5) State: The presence of values like 'AS', 'XX', and 'MP' might indicate incorrect or non-standard state codes.'AS': This likely refers to American Samoa, a U.S. territory in the South Pacific.'XX': This could potentially indicate missing or placeholder data, as 'XX' is not a recognized state abbreviation.'MP': This is likely referring to the Northern Mariana Islands, another U.S. territory in the Western Pacific.These values might represent areas that aren't part of the standard 50 states, which could indicate data entry errors or a need for special handling due to the non-standard state codes. It's a good idea to investigate further to ensure data accuracy and reliability.

6) NAICSCode: The presence of some codes like '999990.0' might indicate missing or improperly classified NAICS codes.

7) BusinessType: The presence of 'nan' values might indicate missing data in the BusinessType column.

8) RaceEthnicity: 'Unanswered' seems to be the dominant category, which could suggest missing data in this column.

9) Gender: 'Unanswered' is the dominant category here as well, which might indicate missing gender data.

10) Veteran: 'Unanswered' again dominates, which suggests missing veteran status data.

11) NonProfit: Most entries are 'Y', but there are also 'nan' values, indicating potential missing data or inconsistencies.

12) DateApproved: There is a wide range of approval dates, which might need to be checked for consistency and possible errors.

13) Lender: Some lenders appear to have various entries (e.g., 'Bank of America, National Association' vs. 'Bank of America'), which might require standardization.

14) CD (Congressional District): There are entries like 'AL -', 'NC -', 'TX -', etc., that seem to have incomplete or improperly formatted codes.
